```python
import pickle
import pandas as pd
```

```python
with open("diarize.pkl", "rb") as file:
    audio_output = pickle.load(file)
```

```python
audio_output
```

```python
dataframe_items = []
for file_name, diarize in audio_output.items():
    for speaker_key, speaker_timeline in diarize.items():
        for item in speaker_timeline:
            dataframe_items.append({
                "filename": file_name,
                "speaker_key": speaker_key,
                "segment_start": item[0],
                "segment_end": item[1]
            })
            
df = pd.DataFrame(dataframe_items)
df
```

```python
df.to_excel("Audio_Diarization.xlsx", index=False)
```
